```python
def determine_marker_color(class_value):
    if class_value == 1:
        return 'green'
    elif class_value == 0:
        return 'red'
    else:
        return 'grey'  # Default color if the class is neither 0 nor 1

# Apply the function to create the 'marker_color' column
spacex_df['marker_color'] = spacex_df['class'].apply(determine_marker_color)

# Display the updated DataFrame
print(spacex_df)
```

```python
import numpy as np
launch_sites_df = pd.DataFrame({
    'Launch Site': ['CCAFS LC-40', 'VAFB SLC-4E', 'KSC LC-39A', 'CCAFS SLC-40'],
    'Lat': [28.562302, 34.632834, 28.573255, 28.563197],
    'Long': [-80.577356, -120.610745, -80.646895, -80.576820]
})

# Haversine function to calculate distance between two points
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of the Earth in kilometers
    phi1 = np.radians(lat1)
    phi2 = np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)
    
    a = np.sin(delta_phi / 2)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    
    distance = R * c
    return distance

# Create an empty DataFrame to store distances
distance_matrix = pd.DataFrame(index=launch_sites_df['Launch Site'], columns=launch_sites_df['Launch Site'])

# Calculate distances between each pair of launch sites
for i, row1 in launch_sites_df.iterrows():
    for j, row2 in launch_sites_df.iterrows():
        if i < j:  # Avoid redundant calculations
            distance = haversine(row1['Lat'], row1['Long'], row2['Lat'], row2['Long'])
            distance_matrix.loc[row1['Launch Site'], row2['Launch Site']] = distance
            distance_matrix.loc[row2['Launch Site'], row1['Launch Site']] = distance

# Convert distance_matrix to numeric values (fill NaNs with 0 or some other value if needed)
distance_matrix = distance_matrix.astype(float)

print(distance_matrix)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
